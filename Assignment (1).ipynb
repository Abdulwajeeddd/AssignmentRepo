```python
import random
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def is_valid_position(x, y, width, height, bin_width, bin_height, positions, rectangles):
    if x < 0 or y < 0 or x + width > bin_width or y + height > bin_height:
        return False
    for (other_x, other_y), (other_width, other_height) in zip(positions, rectangles):
        if other_x is None or other_y is None:
            continue
        if not (
            x + width + 1 <= other_x or
            x >= other_x + other_width + 1 or
            y + height + 1 <= other_y or
            y >= other_y + other_height + 1
        ):
            return False
    return True

def place_rectangles(bin_width, bin_height, num_rectangles):
    rectangles = []
    positions = [(None, None)] * num_rectangles

    for i in range(num_rectangles):
        width = random.randint(5, 15)
        height = random.randint(5, 15)
        rectangles.append((width, height))

        for _ in range(1000):
            x = random.randint(0, bin_width - width)
            y = random.randint(0, bin_height - height)
            if is_valid_position(x, y, width, height, bin_width, bin_height, positions, rectangles):
                positions[i] = (x, y)
                break
        else:
            raise ValueError("Could not place all rectangles with the required spacing.")

    return rectangles, positions

def plot_rectangles(bin_width, bin_height, rectangles, positions):
    fig, ax = plt.subplots()
    ax.set_xlim(0, bin_width)
    ax.set_ylim(0, bin_height)
    ax.set_aspect('equal')

    for i, (width, height) in enumerate(rectangles):
        x, y = positions[i]
        rect_patch = patches.Rectangle((x, y), width, height, linewidth=1, edgecolor='blue', facecolor='lightblue')
        ax.add_patch(rect_patch)
        ax.text(x + width / 2, y + height / 2, str(i + 1), ha='center', va='center')

    plt.title("Rectangle Placement in Bin")
    plt.xlabel("Width")
    plt.ylabel("Height")
    plt.show()


bin_width = 80
bin_height = 40
num_rectangles = 9

rectangles, positions = place_rectangles(bin_width, bin_height, num_rectangles)
plot_rectangles(bin_width, bin_height, rectangles, positions)
```
